```python
#This function separates each molucule to a mol2 file

def mol2tofiles (mol2, mols) :
    """ mol2 : .mol2 file with absolute path
        mols : Absolute path for molecules to save
    """
    newline = '\n'
    delimiter = '@<TRIPOS>MOLECULE'
    with open (mol2, 'r') as f :
        ligands = f.read().split(delimiter) [1:]
    ligands_modified = [delimiter + lig for lig in ligands]
    k = 1
    for ligand in ligands_modified :
        with open (f"{mols}/{str(k)}-{ligand.split(newline)[1]}.mol2" ,'w') as writer :
            writer.write (ligand)
        print (f'{k} Molecules Done!')
        k += 1
```
